```python
# drop columns 
for i in df:
    if(df[i].nunique() == 1):
        print(i)
```

```python
df.drop(['EmployeeCount','Over18','StandardHours'],axis=1,inplace=True)
df.drop(['PerformanceRating'],axis=1,inplace=True)
```

### Feature Engineering

```python
# لفئات Monthly Income تقسيم ال
df['Income_Category'] = pd.cut(df['MonthlyIncome'], 
                            bins=[0, 5000, 10000, 20000], 
                            labels=['Low (<5k)', 'Medium (5k-10k)', 'High (>10k)'])
```

```python
# تقسيم العمر ل categories 
df['Age_Category'] = pd.cut(df['Age'], 
                         bins=[0, 30, 45, 60], 
                         labels=['Young (<30)', 'Middle-Aged (30-45)', 'Senior (>45)'])
```

```python
# Attrition Encoding (yes:1 , No:0 )
df['Encoded_Attrition']=df['Attrition'].map({'Yes':1,'No':0})
# OverTime Encoding 
df['Encoded_OverTime']=df['OverTime'].map({'Yes':1,'No':0})
# Gender Encoding
df['Encoded_Gender']=df['Gender'].map({'Male':1,'Female':0})
# Business Travel Encoding
df['Encoded_BusinessTravel']=df['BusinessTravel'].map({'Non-Travel':0,'Travel_Rarely':1,'Travel_Frequently':2})
```

```python
# yearsWithCurrentManager Binning
df['new_YearsWithCurrManager'] = pd.cut(df['YearsWithCurrManager'],
						   bins=[-1,2,6,20],
						   labels=['0-2 Years','3-6 years','+7 years'])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. اختيار الأعمدة الرقمية فقط (لتجنب الخطأ اللي ظهرلك)
numeric_df = df.select_dtypes(include=[np.number])

# 2. حساب مصفوفة الارتباط
corr_matrix = numeric_df.corr()

# 3. إعداد شكل الرسمة (Heatmap)
plt.figure(figsize=(15, 10))

# رسم الـ Heatmap مع تخصيص الألوان
# cmap='RdBu_r' بيخلي الأحمر للموجب والأزرق للسالب (أو العكس حسب الذوق)
# annot=True عشان يكتب الأرقام جوه المربعات
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', center=0)

plt.title('Correlation Matrix for HR Analytics Variables', fontsize=16)
plt.show()
```

### Multivariate Analysis Analysis

```python
# كود هيقولك نسبة الاستقالة لكل ليفل وظيفي مقسومين بالـ Overtime
pivot_table = df.pivot_table(index='JobLevel', columns='OverTime', values='Encoded_Attrition', aggfunc='mean') * 100
print(pivot_table)
```
